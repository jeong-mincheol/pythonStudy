```python
# 3. 노래제목을 검색하면 SONGID값을 가져오는 로직 만들기

import requests  # requests 파이썬에서 인터넷에서 접근 할 때 사용
from bs4 import BeautifulSoup
import os
import re

searchWrd = input("검색할 노래의 제목을 입력하세요.")

def searchSong(songName):

    url = f"https://www.melon.com/search/total/index.htm?q={songName}&section=&searchGnbYn=Y&kkoSpl=Y&kkoDpType=&mwkLogType=T"
    head = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36'}
    r = requests.get(url, headers=head)

    bs = BeautifulSoup(r.text)
    target = str(bs.find("div", class_="tb_list d_song_list songTypeOne").find_all("tr")[1].find("button","btn_icon play" ))

    p = re.compile("playSong\(\'([0-9]+)\',([0-9]+)\)")
    result = p.findall(target)[0][1]

    return result

def save_lyrics(songIdValue):

    url = f"https://www.melon.com/song/detail.htm?songId={songIdValue}"
    head = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36'}
    r = requests.get(url, headers=head)

    bs = BeautifulSoup(r.text)
    lyrics = BeautifulSoup(str(bs.find("div", id="d_video_summary")).replace("<br/>", "\n")).text

    title = bs.find("div", class_="song_name").text.replace("곡명", "").strip()
    artist = bs.find("div", class_="artist").text.strip()
    if not os.path.isdir("./lyrics/" + artist):
        os.mkdir("./lyrics/" + artist)
    f = open(f"./lyrics/{artist}/{artist}-{title}.txt", 'w', encoding='utf-8')
    f.write(lyrics.strip())
    f.close()

songId = ""
songId = searchSong(searchWrd)
print(songId)
save_lyrics(songId)
```

```python
# 경로 및 URL 인코딩 기능 추가
import requests  # requests 파이썬에서 인터넷에서 접근 할 때 사용
from bs4 import BeautifulSoup
import os
import re
from urllib.parse import urlparse, urlencode, quote

searchWrd = input("검색단어를 입력하세요.")

q= "아이유"

def searchSong(songName):

    payload = {
        'mwkLogType' : ['T'],
        'searchGnbYn' : ['Y'],
        'kkoSpl' : ['Y']
    }
    payload['q'] = [songName]

    url = "https://www.melon.com/search/total/index.htm?" + urlencode(payload, doseq=True)
    head = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36'}
    r = requests.get(url, headers=head)

    bs = BeautifulSoup(r.text)
    target = str(bs.find("div", class_="tb_list d_song_list songTypeOne").find_all("tr")[1].find("button","btn_icon play" ))

    p = re.compile("playSong\(\'([0-9]+)\',([0-9]+)\)")
    result = p.findall(target)[0][1]

    return result

def save_lyrics(songIdValue):

    url = f"https://www.melon.com/song/detail.htm?songId={songIdValue}"
    head = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36'}
    r = requests.get(url, headers=head)

    bs = BeautifulSoup(r.text)
    lyrics = BeautifulSoup(str(bs.find("div", id="d_video_summary")).replace("<br/>", "\n")).text

    title = bs.find("div", class_="song_name").text.replace("곡명", "").strip()
    artist = bs.find("div", class_="artist").text.strip()
    if not os.path.isdir("./lyrics/" + artist):
        os.mkdir("./lyrics/" + artist)
    f = open(f"./lyrics/{artist}/{artist}-{title}.txt", 'w', encoding='utf-8')
    f.write(lyrics.strip())
    f.close()

songId = ""
songId = searchSong(searchWrd)
print(songId)
save_lyrics(songId)
```
